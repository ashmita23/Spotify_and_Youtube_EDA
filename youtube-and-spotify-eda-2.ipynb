```python
columns_to_impute = ['Danceability','Energy','Key','Loudness','Speechiness','Instrumentalness','Acousticness','Liveness','Valence','Tempo']

# Impute missing values with median for the specified columns
for col in columns_to_impute:
    median_value = music_df[col].median()
    music_df[col].fillna(median_value, inplace=True)
```

### Top 10 songs with highest streams , Views and Likes resp

```python
la=music_df['Stream'].sort_values(ascending=False).head(10)
top_10 = music_df[music_df['Stream'].isin(la)]
top_10[['Artist','Views','Stream']]
```

```python
la=music_df['Views'].sort_values(ascending=False).head(10)
views_10= music_df[music_df['Views'].isin(la)]
```

```python
la=music_df['Likes'].sort_values(ascending=False).head(10)
likes_10 = music_df[music_df['Likes'].isin(la)]
```

```python
la=music_df['Comments'].sort_values(ascending=False).head(10)
comments_10=music_df[music_df['Comments'].isin(la)]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

# Top 10 songs with most views
views_10.plot(kind='barh', x='Track', y='Views', ax=axs[0],legend=False)
axs[0].set_title('Top 10 songs with most views')

# Top 10 songs with most comments
comments_10.plot(kind='barh', x='Track', y='Comments', ax=axs[1],legend=False)
axs[1].set_title('Top 10 songs with most comments')

# Top 10 songs with most likes
likes_10.plot(kind='barh', x='Track', y='Likes', ax=axs[2],legend=False)
axs[2].set_title('Top 10 songs with most likes')

fig.tight_layout()


plt.show()
```
